```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.head()
```

```python
cat_cols = df.select_dtypes(include="object")
```

```python
for i in cat_cols:
    df[i] = le.fit_transform(df[i])
df.head()
```

```python
df.info()
```

```python
x = df.drop("price_rupiah", axis = 1)
y = df["price_rupiah"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lm = LinearRegression()
lm.fit(x_train,y_train)

svm = SVR()
svm.fit(x_train,y_train)

rf = RandomForestRegressor()
rf.fit(x_train, y_train)

gbr = GradientBoostingRegressor()
gbr.fit(x_train, y_train)

xg = XGBRegressor()
xg.fit(x_train,y_train)




```

```python
y_pred = lm.predict(x_test)
y_pred2 = svm.predict(x_test)
y_pred3 = rf.predict(x_test)
y_pred4 = xg.predict(x_test)
y_pred5 = gbr.predict(x_test)
```

```python
print(f"Linear Regression {lm.score(x_test,y_test) * 100}")
print(f"SVR {svm.score(x_test,y_test) * 100}")
print(f"Randon Forest {rf.score(x_test,y_test) * 100}")
print(f"XGBoost {xg.score(x_test,y_test) * 100}")
print(f"Gradient Boost {gbr.score(x_test,y_test) * 100}")
```
